```python
def similar_replace(df):
    # Iterate over the DataFrame to replace null values where Pregnancy is False
    for index, row in df.iterrows():
        if not row['Pregnancy status'] and (pd.isnull(row['Height']) or pd.isnull(row['Weight'])):
            # Ensure Age is not null for the current row
            if pd.notnull(row['Age']):
                # Find similar records: same sex and exact age, excluding the current record
                similar_mask = (df['Sex'] == row['Sex']) & \
                               (df['Age'] == row['Age']) & \
                               (df.index != index)

                # Calculate replacement values as the mean of similar records
                height_replacement = df.loc[similar_mask, 'Height'].mean() if pd.isnull(row['Height']) else row['Height']
                weight_replacement = df.loc[similar_mask, 'Weight'].mean() if pd.isnull(row['Weight']) else row['Weight']

                # Replace null values with the calculated replacements
                df.at[index, 'Height'] = height_replacement
                df.at[index, 'Weight'] = weight_replacement

    return df
```

```python
new_dataset = similar_replace(droped_train)
```

```python
#Checks if the pregnancy is true and automattically assigns an Adolescent variable to it
new_dataset.loc[new_dataset["Pregnancy status"] == 1,"Age"] = "Adolescent"
new_dataset.head(100)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

def cramers_v(chi2, n, k1, k2):
    """Calculate Cramr's V statistic for a given Chi2 and sample size."""
    return np.sqrt(chi2 / n / min(k1 - 1, k2 - 1))

# Assuming df is your DataFrame
# Convert 'Pregnancy status' to a string to treat it as a categorical variable
new_dataset['Pregnancy status'] = new_dataset['Pregnancy status'].astype(str)

# Chi-Square Test for Pregnancy status vs. Murmur
contingency_table_murmur = pd.crosstab(new_dataset['Pregnancy status'], new_dataset['Murmur'])
chi2_murmur, p_murmur, _, _ = chi2_contingency(contingency_table_murmur)
cramers_v_murmur = cramers_v(chi2_murmur, new_dataset.shape[0], len(contingency_table_murmur.index), len(contingency_table_murmur.columns))

# Chi-Square Test for Pregnancy status vs. Outcome
contingency_table_outcome = pd.crosstab(new_dataset['Pregnancy status'], new_dataset['Outcome'])
chi2_outcome, p_outcome, _, _ = chi2_contingency(contingency_table_outcome)
cramers_v_outcome = cramers_v(chi2_outcome, new_dataset.shape[0], len(contingency_table_outcome.index), len(contingency_table_outcome.columns))

print(f"Pregnancy status vs. Murmur: Chi2={chi2_murmur}, p={p_murmur}, Cramr's V={cramers_v_murmur}")
print(f"Pregnancy status vs. Outcome: Chi2={chi2_outcome}, p={p_outcome}, Cramr's V={cramers_v_outcome}")
```
